```python
def terror(url):
    
    browser.get(url)
    html =browser.page_source
    
    
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html,'html.parser')
    table=soup.find_all("table",class_='wikitable')

    rows=table[0].find_all("tr")
    
    Attack_type=[]
    no_of_ppl_dead=[]
    Injured=[]
    Location=[]
    Perpetrator=[]
    
    for i in rows[1:]:
        Attack_type.append(i.find_all("td")[1].text.strip())
        no_of_ppl_dead.append(i.find_all("td")[2].text.strip())
        Injured.append(i.find_all("td")[3].text.strip())
        Location.append(i.find_all("td")[4].text.strip())
        Perpetrator.append(i.find_all("td")[6].text.strip())
        
    import pandas as pd
    
    F_table=pd.DataFrame({"Attack_type":Attack_type,"no_of_ppl_dead":no_of_ppl_dead,"Injured":Injured,"Perpetrator":Perpetrator,"Location":Location})
    F_table["url"]=url
    return (F_table)
```

```python
import re
```

```python
pattern=re.compile(r'\d{1,4}')
```

```python
import numpy as np
def get_numbers(x):
    match=re.findall(pattern,x)
    if len(match)==0:
        return np.nan
    elif len(match)>1:
        return float(match[-1])
    else:
        return float(match[0])  
```

```python
big_table['no_of_ppl_dead_cleaned']=big_table['no_of_ppl_dead'].map(get_numbers)
```

```python
big_table.groupby('month')['no_of_ppl_dead_cleaned'].mean()
```

```python
pattern=re.compile(r'\d{1,3}')
```

```python
import numpy as np
def get_numb(x):
    match=re.findall(pattern,x)
    if len(match)==0:
        return np.nan
    elif len(match)==1:
        return float(match[0])
    else:
        match=np.array(match,dtype='float')
        return match.mean()
    
    
    
```

```python
big_table["Injured_c"]= big_table["Injured"].map(get_numb)
```

```python
big_table.groupby("month")["Injured_c"].mean().round(2)
```
